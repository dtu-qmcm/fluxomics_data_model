```python
import os
import incawrapper
```

```python
data_folder = "H:/La meva unitat/Estudis/Master Biotech/Thesis/fluxomics/ecolimodel/output"
output_files = os.listdir(data_folder)
out_folder = "H:/La meva unitat/Estudis/Master Biotech/Thesis/fluxomics/ecolimodel/output/csv_data"
fluxes = {}
for exp_data in output_files:
    res = incawrapper.INCAResults(os.path.join(data_folder, exp_data))
    print(exp_data)
    print(res.fitdata.get_goodness_of_fit())
    fluxes[exp_data] = res.fitdata.fitted_parameters
    fluxes[exp_data].to_csv(os.path.join(out_folder, exp_data.split(".")[0] + ".csv"))
```
